```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calculate_vif_(X):

    '''X - pandas dataframe'''
    thresh = 5.0
    variables = range(X.shape[1])

    for i in np.arange(0, len(variables)):
        vif = [variance_inflation_factor(X[variables].values, ix) for ix in range(X[variables].shape[1])]
        print(vif)
        maxloc = vif.index(max(vif))
        if max(vif) > thresh:
            print('dropping \'' + X[variables].columns[maxloc] + '\' at index: ' + str(maxloc))
            del variables[maxloc]

    print('Remaining variables:')
    print(X.columns[variables])
    return X
```

```python
df=df[['Balance',
    'EstimatedSalary',
    'Age',
    'IsActiveMember',
    'CreditScore',
    'Gender',
    'Geography','Exited']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import RFECV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(df.iloc[:,:-1],df.iloc[:,-1],test_size=0.2,shuffle=True,random_state=42)
```

```python
lr=LogisticRegression(random_state=0)
rfc=RandomForestClassifier(n_estimators=500,random_state=0)
dtc=DecisionTreeClassifier(random_state=0)
xg=XGBClassifier(n_estimators=500)
```

```python
print('logistic regression')
select_1=RFECV(lr,step=1,cv=5).fit(X_train,Y_train)
pd.concat([pd.Series(np.array(df.columns)[:-1]),pd.Series(select_1.ranking_),pd.Series(select_1.support_)],axis=1)
```

```python
print('randomforest')
select_2=RFECV(rfc,step=1,cv=5).fit(X_train,Y_train)
pd.concat([pd.Series(np.array(df.columns)[:-1]),pd.Series(select_2.ranking_),pd.Series(select_2.support_)],axis=1)
```

```python
print('decisiontree')
select_3=RFECV(dtc,step=1,cv=5).fit(X_train,Y_train)
pd.concat([pd.Series(np.array(df.columns)[:-1]),pd.Series(select_3.ranking_),pd.Series(select_3.support_)],axis=1)
```

```python
print('xgboost')
select_4=RFECV(xg,step=1,cv=5).fit(X_train,Y_train)
pd.concat([pd.Series(np.array(df.columns)[:-1]),pd.Series(select_4.ranking_),pd.Series(select_4.support_)],axis=1)
```

```python
result=rfc.fit(X_train,Y_train)
```

```python
result.score(X_train,Y_train)
```

```python
y_predict=result.predict(X_test) 
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_predict,Y_test)
```
